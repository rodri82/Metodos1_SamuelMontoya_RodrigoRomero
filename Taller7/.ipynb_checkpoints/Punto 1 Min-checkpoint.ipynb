```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
A=np.zeros((3,2))
A[0][0]=2
A[0][1]=-1
A[1][0]=1
A[1][1]=2
A[2][0]=1
A[2][1]=1
b=np.array([2,1,4])
AAt=np.dot(A.T,A)
Atb=np.dot(A.T,b)
sol=np.linalg.solve(AAt,Atb)
sol
```

```python
f=lambda x: 2*x-2
x=np.linspace(-5,8,100)
y_=f(x)
```

```python
f2=lambda x:(1-x)/2
x2=np.linspace(-5,8,100)
y2_=f2(x)
```

```python
f3=lambda x:4-x
x3=np.linspace(-5,8,100)
y3_=f3(x)
```

```python
plt.plot(x,y_)
plt.plot(x2,y2_)
plt.plot(x3,y3_)
plt.plot(sol[0],sol[1],"o")
plt.xlim(0,8)
plt.ylim(-5,6)
plt.show
```

```python
def Distancia(A,x,y,b):
    lis=[]
    for i in range(len(x)):
        minlis=[]
        for j in range(len(y)):
            Ax=np.dot(A,[x[i],y[j]])
            minlis.append(np.linalg.norm(Ax-b))
        lis.append(minlis)
    return np.array(lis)
```

```python
x=np.linspace(-5,5,333)
y=np.linspace(-5,5,333)

X,Y = np.meshgrid(x,y)
dist=Distancia(A,x,y,b)
fig,ax = plt.subplots(subplot_kw={"projection":"3d"})
surf = ax.plot_surface(X,Y,dist,cmap=cm.coolwarm)
```

```python
menor=100000
coor=[0,0]
for i in range(len(dist)):
    for j in range(len(dist[0])):
        if dist[i][j]<menor:
            menor=dist[i][j]
            coor=[x[i],y[j]]
print(menor,coor)

a=np.min(dist)
```
