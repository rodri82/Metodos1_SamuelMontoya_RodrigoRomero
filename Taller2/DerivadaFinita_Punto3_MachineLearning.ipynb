```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
h = 0.05
x = np.linspace(-10,10,int(20/h))
f = lambda x: 1/np.sqrt(1+np.exp(-x**2))
y = f(x)
```

```python
def derivada(f,x,h):
    M = np.array([1, 0, -1])
    fx = f(x)
    D = np.sum(np.multiply(M,np.transpose(np.array([[f(x+h)],[f(x)],[f(x-h)]]))),axis=2)/(2*h)
    return D
```

```python
df = derivada(f,x,h)

fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x, y, color='g', label='f(x)')
ax1.plot(x,df,color='r',label='Derivada num√©rica')
ax1.legend(loc=0)
ax1.grid(alpha=0.25)
```
