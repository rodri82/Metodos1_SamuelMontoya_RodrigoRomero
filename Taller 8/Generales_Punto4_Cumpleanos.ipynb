```python
import numpy as np
import matplotlib.pyplot as plt
```

Sea n el n√∫mero de personas

Casos posibles = $365^n$

Casos favorables = 365 * 364 * 363 * ... * (365-n+1) = $\frac{365!}{(365-n)!}$

Probabilidad = $\frac{ Casos Favorables }{ Casos Totales } = \frac{365!}{365^n (365-n)!}$

```python
def GetProbability(n):
    P = (np.math.factorial(365))/((365**n)*(np.math.factorial(365-n)))
    return P
```

```python
Probs = np.zeros(80)
for i in range(1,81):
    Probs[i-1] = GetProbability(i)
n = np.arange(1,81,1)
```

```python
plt.plot(n,Probs,'r+')
plt.title('Probabilidad de que n personas tengan cumplea√±os distintos')
plt.xlabel('Personas')
plt.ylabel('P')
plt.grid()
```
